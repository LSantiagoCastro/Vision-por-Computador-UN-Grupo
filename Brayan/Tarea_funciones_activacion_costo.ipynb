# **Funciones de acativacion**

La neurona artificial genera la salida con base en el proceso de las señales de entrada, esto lo hace gracias a las funciones de activacion. Por lo tento es de vital importancia conocer los diferentes tipos de funciones de activacion que se tienen 



## **Identity**

**Ecuacion:**

 $f(x) = x$

 $f: ℝ →ℝ$

**Rango:** $(-∞,∞)$
 
 **Continuidad:** $C∞$

 **Derivada:**  $f'(x) = 1$


Con la función de activación Identidad, la entrada de un nodo es igual a la salida. Es perfecto para tareas donde el comportamiento subyacente es lineal (similar a la regresión lineal). Cuando no hay linealidad, usar la función de activación por sí sola no es suficiente, pero aún se puede usar como una función de activación en el nodo de salida final para las tareas de regresión.

## **Step**

**Ecuacion**

$
f(x) = \left\lbrace
\begin{array}{ll}
1 \ \ \ para \ \ \ x \ge 0\\
0 \ \ \ para \ \ \ x < 0
\end{array}
\right.
$

**Rango:** (0,1)

**Derivada**

$
f'(x) = \left\lbrace
\begin{array}{ll}
0 \ \ \ para \ \ \ x \not= 0\\
? \ \ \ para \ \ \ x = 0
\end{array}
\right.
$


Esta función también conocida como escalón, indica que si la x es menor que cero la neurona va a ser cero pero cuando es mayor igual a cero dará como salida igual 1. Esta función se usa cuando se quiere clasificar o cuando se tiene salidas categóricas. Por ejemplo, se puede usar para predecir si compro algo o no.

## **Sigmoid**

**Ecuacion**

$f(x) = \frac{1}{1+e^{-x}}$


**Rango:** (0.1)

**Continuidad:** $C∞$

**Derivada**

$f'(x) = \frac{e^{-x}}{(1+e^{-x})^2}$

$=f(x)(1-f(x))$

**Rango de la derivada** (0,0.25)



Sigmoid es bien conocido por su importancia en la regresión logística, con un rango de valores de 0 a 1. La función de activación logística sigmoidea (o sigmoidea como se le llama comúnmente) introduce el concepto de probabilidad en las redes neuronales. Su derivada es distinta de cero y fácil de calcular (es una función de su salida inicial). Sin embargo, en la tarea de clasificación, sigmoide se reemplaza gradualmente por la función Tanh como la función de activación estándar, porque esta última es una función extraña (simétrica respecto al origen).

## **Tanh**


**Ecuacion**

$f(x) = tanh(x)$

$= \frac{2}{1+e^{2x}}-1$

**Rango:**(-1,1)

**Derivada**

$f'(x) = 1-tanh^2(x)$


En la tarea de clasificación, la función tangente hiperbólica (Tanh) reemplaza gradualmente a la función Sigmoide como la función de activación estándar, que tiene muchas características favorecidas por las redes neuronales. Es completamente diferenciable, antisimétrico, y el centro de simetría está en el origen. Para resolver el problema del aprendizaje lento y / o la desaparición gradual, se puede utilizar una variación más gradual de esta función (log-log, softsign, sigmoide simétrico, etc.)

## **ReLu**

**Ecuacion**

$
f(x) = \left\lbrace
\begin{array}{ll}
x \ \ \ para \ \ \ x \ge 0\\
0 \ \ \ para \ \ \ x < 0
\end{array}
\right.
$

**Rango:** $[0,∞)$

**Derivada**

$
f(x) = \left\lbrace
\begin{array}{ll}
1 \ \ \ para \ \ \ x \ge 0\\
0 \ \ \ para \ \ \ x < 0
\end{array}
\right.
$

La unidad lineal rectificada (ReLU) es la función de activación más utilizada en las redes neuronales. Conserva la inspiración biológica de la función escalonada (la neurona solo se activa cuando la entrada supera el umbral), pero cuando la entrada es positiva, la derivada no es cero, lo que permite el aprendizaje basado en gradiente (aunque cuando x = 0, la derivada No está definido). El uso de esta función puede hacer que el cálculo sea muy rápido, porque ni la función ni su derivada contienen operaciones matemáticas complejas. Sin embargo, cuando la entrada es negativa, la velocidad de aprendizaje de ReLU puede volverse muy lenta, o incluso hacer que la neurona sea directamente inválida, porque la entrada es menor que cero y el gradiente es cero, por lo que su peso no puede actualizarse. El proceso de capacitación permanecerá en silencio.

#Etropia cruzada 

La función de pérdida de entropía cruzada es una función de optimización que se utiliza para entrenar modelos de clasificación que clasifican los datos prediciendo la probabilidad (valor entre 0 y 1) de que los datos pertenezcan a una clase u otra. En caso de que la probabilidad predicha de la clase sea muy diferente de la etiqueta de clase real (0 o 1), el valor de la pérdida de entropía cruzada es alto. En caso de que la probabilidad predicha de la clase esté cerca de la etiqueta de clase (0 o 1), la pérdida de entropía cruzada será menor. La pérdida de entropía cruzada se utiliza habitualmente como función de pérdida para los modelos que tienen una salida softmax. Recordemos que la función softmax es una generalización de la regresión logística a múltiples dimensiones y se utiliza en la regresión logística multinomial. 

En particular, la pérdida de entropía cruzada o la función de pérdida logarítmica se utiliza como función de coste para los modelos de regresión logística o los modelos con salida softmax (regresión logística multinomial o red neuronal) con el fin de estimar los parámetros. # Hinge loss
EN el aprendizaje automático, hinger loss es una función de perdida
utiliza para la clasificación de "margen máximo", sobre todo para las máquinas de vectores de soporte (SVM).

Esta función de pérdida se utiliza si los valores objetivo están en el conjunto (-1, 1). La variable objetivo debe modificarse para que tenga valores en el conjunto (-1, 1), lo que significa que si y tiene valor como 0, debe cambiarse como -1.

La función de pérdida se define como

$\ell(y) = \max(0, 1-t \cdot y)$


La función hinge loss trata de asegurar el signo correcto asignando más error si hay una diferencia en el signo entre los valores de clase reales y los predichos.

```python
#Funciones de activacion

class Activation_fun():

  """
  Class of activation functions. The following functions can be used:
  Identity
  step
  sigmoid
  tanh
  ReLu
  """
  import numpy as np
  def __init__(self):
    pass

  def identity(y):
    """
    Activation function identity, input parameters
    y = array
    outuput 
    y = array
    """
    return y

  def step( y):
    """
    Activation function step, input parameters
    y = array
    output 
    newy = array 
    
    """
    newy = np.asarray(y)
    for i in range(len(y)):
      if y[i] >= 0:
        newy[i]=1       
      else:newy[i]=0

    return newy


  def sigmoid(y):
    """
    Activatio function sigmoid, input parameters
    y = array
    output 
    sig 
    """
    import math
    ny = []
    for num in y:
        ny.append(1/(1 + math.exp(-num)))
    return ny
  

   



  def tanh(y):
    """
    Activation function hyperbolic tangent, input parameters
    y = array
    output
    th    
    """

    th =(np.exp(y)-np.exp(-y))/(np.exp(y)+np.exp(-y))
    return th

  def ReLU(y):
    """
    Activation function ReLu, input parameters
    y = array
    output
    rl
    """
    rl = y * (y > 0)
    return rl
  
 
```
