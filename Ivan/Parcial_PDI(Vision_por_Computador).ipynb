```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
class imagenes : #crear clase
  """
  Cmap
  matplot
  """
  def __init__(self):
    pass

  def MatplotExpFunct():

    X=np.linspace(0,10,20)
    B = 1- np.exp(-X/2)

    C= X/10
    #a=fig = plt.figure(figsize=(10,5))
    b=plt.plot(X, B, 'bo--', label='exp')
    c=plt.plot(X, C, 'ro--', label='exp')

    d=plt.title('X vs Y')
    e=plt.ylabel('Y')
    f=plt.xlabel('X')
    g=plt.grid()
    h=plt.legend(loc=0)
    return b,c,d,e,f,g,h

  def MatplotTrigFunct():
    #a=fig = plt.figure(figsize=(10,5))
    x = np.linspace(0, 2*np.pi, 100)
    y1 = np.cos(x)
    y2 = np.sin(x)
    b=plt.plot(x, y1, 'bo--', label='cos')
    c=plt.plot(x, y2, 'r*-', label='sin')
    d=plt.title('X vs Y')
    e=plt.ylabel('Y')
    f=plt.xlabel('X')
    g=plt.grid()
    h=plt.legend(loc=0)
    return b,c,d,e,f,g,h

  def Cmap():
    s=401 #size
    X = np.linspace(-3,3,s)
    Y=X.reshape(s,1)
    Z=np.sqrt(X**2+Y**2)
    a=plt.imshow(Z, cmap='hsv', vmin=0, vmax=4.5, extent=(-3., 3., -3., 3.))
    b=plt.colorbar()
    c=plt.title('Z=sqr(X**"+Y**2)')
    d=plt.ylabel('Y')
    e=plt.xlabel('X')
    return a,b,c,d,e

  def Hist():
# create data
    x = np.arange(4)
    y1 = [30, 20, 50, 40]
    y2 = [40, 60, 50, 10]
    y3 = [10, 50, 10, 30]
    width = 0.2
  
# plot data in grouped manner of bar type
    a=plt.bar(x-0.2, y1, width, color='cyan')
    b=plt.bar(x, y2, width, color='orange')
    c=plt.bar(x+0.2, y3, width, color='green')
    d=plt.xticks(x, ['2018', '2019', '2020', '2021'])
    e=plt.xlabel("AÃ±os")
    f=plt.ylabel("Ventas")
    g=plt.legend(["Audi", "BMW", "Chevrolet"])
    return a,b,c,d,e,f,g
```

```python
img=imagenes
grafico1=img.MatplotExpFunct
grafico2=img.MatplotTrigFunct
grafico3=img.Cmap
grafico4=img.Hist
```

```python
import matplotlib.pyplot as plt 

fig=plt.figure(figsize=(20,9))

ax_1=fig.add_subplot(241)
grafico4()

ax_2=fig.add_subplot(245)
grafico2()

ax_3=fig.add_subplot(142)
grafico1()

ax_3=fig.add_subplot(122)
grafico3()

fig.suptitle("Figure with multiple Subplots")
plt.show()
```
